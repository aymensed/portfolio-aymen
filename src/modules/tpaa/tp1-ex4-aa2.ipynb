```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Données fournies
data = np.array([(1, 2), (2, 5), (3, 10), (4, 17), (5, 26)])
X = data[:, 0].reshape(-1, 1)  # Extraire les valeurs de x
y = data[:, 1]  # Extraire les valeurs de y

# Créer un modèle polynomial de degré 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajuster le modèle de régression linéaire sur les données polynomiales
model = LinearRegression()
model.fit(X_poly, y)

# Prédire les valeurs de y
y_pred = model.predict(X_poly)

# Afficher les coefficients du polynôme
print(f"Coefficients du polynôme : {model.coef_}")
print(f"Ordonnée à l'origine : {model.intercept_}")

# Tracer les points de données
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Données réelles')

# Tracer la courbe du polynôme
X_range = np.linspace(1, 5, 100).reshape(-1, 1)
X_range_poly = poly.transform(X_range)
y_range_pred = model.predict(X_range_poly)
plt.plot(X_range, y_range_pred, color='red', label='Courbe du polynôme (degré 2)', linewidth=2)

# Ajouter des étiquettes et un titre
plt.xlabel('x')
plt.ylabel('y')
plt.title('Régression Non Linéaire (Polynôme de degré 2)')
plt.axhline(0, color='black', linewidth=0.5, ls='--')
plt.axvline(0, color='black', linewidth=0.5, ls='--')
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.legend()

# Afficher le graphique
plt.show()
```
