```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge, LinearRegression

# Step 1: Prepare the dataset
X = np.array([[1], [2], [3], [4], [5]])  # Features
y = np.array([2, 4, 6, 8, 10])           # Labels

# Step 2: Fit Ridge Regression
alpha = 0.1
ridge_model = Ridge(alpha=alpha)
ridge_model.fit(X, y)

# Step 3: Fit Simple Linear Regression
linear_model = LinearRegression()
linear_model.fit(X, y)

# Step 4: Generate predictions
X_test = np.linspace(0, 6, 100).reshape(-1, 1)  # Test points for plotting
ridge_pred = ridge_model.predict(X_test)
linear_pred = linear_model.predict(X_test)

# Step 5: Plotting the results
plt.scatter(X, y, color='red', label='Data points')
plt.plot(X_test, ridge_pred, color='blue', label='Ridge Regression (alpha=0.1)')
plt.plot(X_test, linear_pred, color='green', label='Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Comparison of Ridge and Linear Regression')
plt.legend()
plt.grid()
plt.show()
```
