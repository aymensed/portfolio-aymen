```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Données fournies
X1 = np.array([1, 2, 3, 4, 5])
X2 = np.array([5, 4, 3, 2, 1])
y = np.array([1, 2, 3, 4, 5])

# Combiner X1 et X2 dans une seule matrice
X = np.column_stack((X1, X2))

# Créer et ajuster le modèle de régression linéaire multiple
model = LinearRegression()
model.fit(X, y)

# Afficher les coefficients
print(f"Ordonnée à l'origine (Intercept): {model.intercept_}")
print(f"Coefficient pour X1: {model.coef_[0]}")
print(f"Coefficient pour X2: {model.coef_[1]}")

# Créer un graphique 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Tracer les points de données réelles
ax.scatter(X1, X2, y, color='blue', label='Données réelles')

# Créer une grille pour le plan de régression
X1_grid, X2_grid = np.meshgrid(np.linspace(min(X1), max(X1), 10), np.linspace(min(X2), max(X2), 10))
y_grid = model.intercept_ + model.coef_[0] * X1_grid + model.coef_[1] * X2_grid

# Tracer le plan de régression
ax.plot_surface(X1_grid, X2_grid, y_grid, color='red', alpha=0.5, rstride=100, cstride=100)

# Ajouter des étiquettes
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')

# Ajouter un titre
plt.title('Régression Linéaire Multiple')

# Afficher la légende
plt.legend()

# Afficher le graphique
plt.show()
```
